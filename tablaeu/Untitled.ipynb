```python
import pandas as pd
from functools import reduce

df = pd.read_excel('assignment-01-data-unformated.xlsx')
df = df.reset_index(drop=False)

dfl = list()
col = ['soft corals','sea fans','blue corals','hard corals','sea pens']
for x in range(0,5):
    df1 = df.iloc[:,:11]
    df1.columns = df1.iloc[0]
    df1 = df1[1:]
    res = pd.melt(df1, id_vars=['name','longitude','latitude'], var_name=['Year'])
    res = res.rename(columns = {'value':col[x]})
    dfl.append(res)
    df.drop(df.columns[[3,4,5,6,7,8,9,10]], axis=1, inplace=True)
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['name','longitude','latitude','Year'],
                                            how='outer'), dfl)
data = pd.melt(df_merged, id_vars=['name','longitude','latitude','Year'], var_name=['Corals'])
```
